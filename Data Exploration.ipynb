```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
houses = pd.read_csv('train.csv')
```

```python
houses.head()
```

```python
#plotting null values to visually see missing data
sns.heatmap(houses.isnull())
```

```python
#determing what percentage of each column is missing
percent_null = np.sum(houses.isnull())/len(houses)

#finding and removing which columns have a significant amount of missing data
null_threshold_exceeded = percent_null[percent_null>=0.6]
houses.drop(list(null_threshold_exceeded.index), axis=1, inplace=True)

#removes Id column as this serves no predicitve value
houses.drop(['Id'], axis=1, inplace=True)
```

```python
#seperating features into numerical and categorical
num_features = []
cat_features =[]

for feature in list(houses.columns):
    if(houses[feature].dtype == "O"):
        cat_features.append(feature)
    else:
        num_features.append(feature)
```

```python
%matplotlib inline
for feature in num_features:
    plt.hist(houses[feature])
    plt.xlabel(feature)
    plt.show()
```
